### Feature engineering

```python
#SETTING THE 15 ROWS VALUE IN NUMERIC FORM
for i in range (data.shape[0]):
    if data[15][i] == '+':
        data[15][i] = 1
    else:
        data[15][i] = 0
```

```python
# CHECKING VALUE IN DIFFERENT FEATURE
for i in data.columns:
    print(".........%s value........\n"%i)
    print(data[i].value_counts())
```

```python
# replacing '?' in categorical feature with 'Unknown'
def categorical_replace(features_index):
    for i in features_index:
        for j in range(data.shape[0]):
            if data[i][j] == '?':
                data[i][j]  ='Unknown'
list1 = [0,3,4,5,6,8,9,11,12]
categorical_replace(list1)
```

```python
# changing datatype of numeric field from object to int or float
def datatype_change(features_list):
        for i in features_list:
            data[i] = pd.to_numeric(data[i],errors = 'coerce')
list2 = [1,2,7,10,13,14,15]
datatype_change(list2)
```

```python
## Replacing the '?' value in numerical features with mean
def numerical_replace(features_index):
    for i in features_index:
        for j in range(data.shape[0]):
            if data[i][j] == '?':
                data[i][j]  = data[i].mean()
list3 = [1,2,7,10,13,14]
numerical_replace(list3)
```

```python
# normalising the numeric features
def normalisation(features_index):
    for i in features_index:
        min_feature = data[i].min()
        max_feature = data[i].max()
        data[i] = (data[i]-min_feature)/(max_feature-min_feature)
list4 = [1,2,7,10,13,14]
normalisation(list4)
```

```python
from sklearn.preprocessing import LabelEncoder as le
def label_encoding(feature_list):
    for i in feature_list:
        number = le()
        data[i] = number.fit_transform(data[i].astype('str'))
list5 = [0,3,4,5,6,8,9,11,12] 

label_encoding(list5)
```

```python
#APPLIYING PCA TO AVOID OVERFITTING AND UNDERFIITING
x = data.drop(data[15],axis =1)
y = data[15]
pca = PCA(n_components=7)
X_pca = pca.fit_transform(x)
X_pca.shape
```

### Model Building

```python
# splitting trainig and test data
from sklearn.model_selection import train_test_split as tts
train_x,test_x,train_y,test_y = tts(X_pca,y,random_state = 50 ,stratify = y,train_size = 0.7,test_size = 0.3)
```

```python
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

### Cross validation

```python
# K fold cross validation using 10 folds using KNN classifier

cv = KFold(n_splits=10,random_state=1,shuffle=True)
knn_model = knn(n_neighbors = 3)  #object of knn classifier
score_knn  = cvs(knn_model,X = train_x, y=train_y,cv =cv,scoring = "accuracy")
accuracy = score_knn.mean()*100
print("Accurcy on knn model using k cross validation is %d" %accuracy)
```

```python
# k fold cross validation using 10 fold using logistic regression classifier
lr_model = lr()   # object of logistic regression classifier
score_logistic = cvs(lr_model , X = train_x,y = train_y,scoring = 'accuracy',cv = cv)
accuracy1 = score_logistic.mean()*100
print("Accurcy on logistic regression classifier model using k cross validation is %d" %accuracy1)
```

```python
# k fold cross validation using 10 fold using Decision tree classifier
decision_model = DecisionTreeClassifier(max_depth=6) #object of decision tree classifier
score_decision = cvs(decision_model,X = train_x,y=train_y,scoring = 'accuracy',cv = cv)
accuracy2 = score_decision.mean()*100
print("Accurcy on Decision Tree model using k cross validation is %d" %accuracy2)
```

```python
# k fold cross validation using 10 fold using random forest classifier
randomforest_model = RandomForestClassifier(max_depth=6,max_features=5)  # object of random forest classifier
score_forest = cvs(randomforest_model,X = train_x,y=train_y,scoring = 'accuracy',cv = cv)
accuracy4 = score_forest.mean()*100
print("Accurcy on random forest model using k cross validation is %d" %accuracy4)
```

### Models Training and testing  


```python
# MODEL TRAINING AND TESTING OF KNN
knn_model.fit(train_x,train_y)
train_predict = knn_model.predict(train_x)
training_accuracy = accuracy_score(train_y,train_predict)*100
print("Training accuracy of knn model is : {}".format(training_accuracy))
test_predict = knn_model.predict(test_x)
testing_accuracy = accuracy_score(test_y,test_predict)*100
print("Testing accuracy of knn model is : {}".format(testing_accuracy))
training_precision = precision_score(train_y,train_predict, average=None)
print("Trainng precision of knn model is : {}".format(training_precision))
testing_precision = precision_score(test_y,test_predict, average=None)
print("Testing precision of knn model is : {}".format(testing_precision))
```

```python
#Hyperparameter tuning of decision tree
def decision_best():
    train_matrix = []
    test_matrix = []
    depth = [i for i in range(1,10) ]
    for i in depth:
            decision_model = DecisionTreeClassifier(max_depth=i)
            decision_model.fit(train_x,train_y)
            train_predict2 = decision_model.predict(train_x)
            training_accuracy = accuracy_score(train_y,train_predict2)*100
            train_matrix.append(training_accuracy)
            test_predict2 = decision_model.predict(test_x)
            testing_accuracy = accuracy_score(test_y,test_predict2)*100
            test_matrix.append(testing_accuracy)
    frame = pd.DataFrame({'max_depth':range(1,10),
                          'training_accuracy':train_matrix,
                          'testing_accracy': test_matrix})       
    return frame        
Frame_decisiontree = decision_best()
Frame_decisiontree.head(10)
```
